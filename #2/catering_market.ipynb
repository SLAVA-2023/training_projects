## Предобработка данных

Далее выведем информацию о количестве пропусков по столбцам

```python
data.isnull().sum()
```

В пяти столбцах датасета имеется значительное количество пропусков. Подобрать какой-либо способ замены, не искажающий исходные данные не получится, и в дальнейшем может привести к ошибочным выводам.  Оставим пропуски в столбцах без изменений. Если понадобится выполнять какие-либо манипуляции с этими столбцами в дальнейшем, можем заменить пропуски "заглушками". На данном этапе нам этого не требуется.

Также в столбце `hours` имеем 536 пропусков, около 7% от общего количества строк. Не стоит удалять такое количесво данных, рассмотрим подробнее. Часы работы, чаще всего, зависят от категории заведения, поэтому выведем все категории заведений, где встречаются пропуски

```python
data.loc[data['hours'].isnull(), 'category'].value_counts()
```

Получили, что большая часть пропусков связано с кафе. Далее выведем часы работы кафе, представленных в наборе данных.

```python
data.loc[data['category'] == 'кафе', 'hours'].value_counts()
```

```python
data.loc[data['category'] == 'кафе', 'category'].count()
```

Получили, что около 10% кафе работают круглосуточно. Остальные же предпочитают на ночь закрываться. Поэтому для заполнения пропусков, где тип заведения кафе, используем наиболее популярный дневной режим работы - "ежедневно, 10:00–22:00". В остальных случаях удалим пропуски.

```python
data.loc[(data['category'] == 'кафе') & (data['hours'].isnull()), 'hours'] =\
data.loc[(data['category'] == 'кафе') & (data['hours'].isnull()), 'hours'].fillna('ежедневно, 10:00–22:00')
```

```python
data = data.dropna(subset=['hours'])
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

Создаем столбец `street`

```python
data['street'] = data['address'].str.split(pat=', ').str[1]
```

Создаем столбец `is_24/7`

```python
data['is_24/7'] = data['hours'].str.contains('круглосуточно')
```
